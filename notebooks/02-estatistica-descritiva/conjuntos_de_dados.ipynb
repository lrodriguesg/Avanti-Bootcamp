## Estatística de cada variável categórica

### Espécie

```python
# Costruindo tabela de frequência
f_abs_species = df_p['species'].value_counts()
f_rel_species = df_p['species'].value_counts(normalize=True) * 100
dist_f_specie = pd.DataFrame({'Frequência Absoluta': f_abs_species, 'Frequência Relativa (%)': f_rel_species})
dist_f_specie.rename_axis('Espécie', axis= 'columns', inplace = True)
dist_f_specie
```
